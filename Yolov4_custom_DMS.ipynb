```python
import os

# Define the directories
train_dir = '/content/darknet_for_colab/data/train/images/'
valid_dir = '/content/darknet_for_colab/data/valid/images/'
test_dir = '/content/darknet_for_colab/data/test/images/'

# Initialize lists to store image file paths
train_images = []
valid_images = []
test_images = []

# Populate lists with paths to the image files in each directory
for file_name in os.listdir(train_dir):
    if file_name.endswith('.jpg'):
        train_images.append(train_dir + file_name)

for file_name in os.listdir(valid_dir):
    if file_name.endswith('.jpg'):
        valid_images.append(valid_dir + file_name)

for file_name in os.listdir(test_dir):
    if file_name.endswith('.jpg'):
        test_images.append(test_dir + file_name)

# Write the paths to the 'train.txt' file
with open('train.txt', 'w') as file_object:
    for img in train_images:
        file_object.write(img + '\n')

# Write the paths to the 'valid.txt' file (if applicable)
with open('valid.txt', 'w') as file_object:
    for img in valid_images:
        file_object.write(img + '\n')

# Write the paths to the 'test.txt' file
with open('test.txt', 'w') as file_object:
    for img in test_images:
        file_object.write(img + '\n')

print("train.txt, valid.txt, and test.txt have been created successfully.")
```

```python
import os
import shutil
import random

# Define the source directory where all files are currently located
source_dir = '/content/darknet_for_colab/'

# Define the target ts directory where we want to organize the files
ts_dir = '/content/darknet_for_colab/ts/'

# Create directories for train, valid, and test if they don't exist
train_dir = os.path.join(ts_dir, 'train/images')
valid_dir = os.path.join(ts_dir, 'valid/images')
test_dir = os.path.join(ts_dir, 'test/images')

train_labels_dir = os.path.join(ts_dir, 'train/labels')
valid_labels_dir = os.path.join(ts_dir, 'valid/labels')
test_labels_dir = os.path.join(ts_dir, 'test/labels')

# Ensure the directories exist
os.makedirs(train_dir, exist_ok=True)
os.makedirs(valid_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

os.makedirs(train_labels_dir, exist_ok=True)
os.makedirs(valid_labels_dir, exist_ok=True)
os.makedirs(test_labels_dir, exist_ok=True)

# List all image files in the source directory (we'll assume the image files are in the same directory as the labels)
image_files = [f for f in os.listdir(source_dir) if f.endswith('.jpg')]

# Split the image files into train, valid, and test sets
random.shuffle(image_files)  # Shuffle to randomize the split

# You can adjust the split ratios as needed
train_size = int(0.7 * len(image_files))   # 70% for training
valid_size = int(0.2 * len(image_files))   # 20% for validation
test_size = len(image_files) - train_size - valid_size  # Remaining for test

# Split the image files
train_images = image_files[:train_size]
valid_images = image_files[train_size:train_size + valid_size]
test_images = image_files[train_size + valid_size:]

# Function to move images and their corresponding labels
def move_files(image_list, image_dir, label_dir):
    for image in image_list:
        # Move the image file
        shutil.move(os.path.join(source_dir, image), os.path.join(image_dir, image))

        # Move the corresponding label file (assuming the label file has the same name but a .txt extension)
        label_file = image.replace('.jpg', '.txt')
        shutil.move(os.path.join(source_dir, label_file), os.path.join(label_dir, label_file))

# Move files to the appropriate directories
move_files(train_images, train_dir, train_labels_dir)
move_files(valid_images, valid_dir, valid_labels_dir)
move_files(test_images, test_dir, test_labels_dir)

# Create the train.txt, valid.txt, and test.txt files
def write_paths_to_txt(image_list, txt_file):
    with open(txt_file, 'w') as f:
        for image in image_list:
            # Write the full path to the image
            f.write(os.path.join(ts_dir, image.replace('.jpg', '.txt')) + '\n')

# Write paths for train, valid, and test sets
write_paths_to_txt(train_images, 'train.txt')
write_paths_to_txt(valid_images, 'valid.txt')
write_paths_to_txt(test_images, 'test.txt')

print("Directories and .txt files have been created successfully.")
```
